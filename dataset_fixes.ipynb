```python
superset_data = load_dataset("lighteval/MATH")
```

```python
superset_data_train = superset_data['train'].to_pandas()
superset_data_test = superset_data['test'].to_pandas()
```

```python
import pandas as pd
superset_data = pd.concat([superset_data_train, superset_data_test])
```

```python
superset_data.reset_index()
```

```python
import pandas as pd

def createPrompt(row):
    bos_token = '<s>'
    system_prompt = '[INST]<<SYS>> You are an expert mathematician, solve the given question accordingly <</SYS>>'
    input_prompt = f" {row['problem']}[/INST]"
    output_prompt = f"{row['solution']}</s>"

    return bos_token + system_prompt +   input_prompt + output_prompt

# Assuming 'data' is your DataFrame
superset_data['Text'] = superset_data.apply(createPrompt, axis=1)
```
